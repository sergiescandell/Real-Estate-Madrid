##  1. ¿Cuáles son las propiedades que tienen al menos 2 baños, un ascensor y una plaza de aparcamiento? Selecciona el ASSETID, BATHNUMBER, HASLIFT, HASPARKINGSPACE y PRICE 

```python
sSQL =  """ 

SELECT assetid, bathnumber, haslift, hasparkingspace, price
FROM inmuebles
WHERE bathnumber >= 2 
    AND HASLIFT = 1 
    AND HASPARKINGSPACE = 1

"""
pd.read_sql(sSQL,conn)
```

## 2. ¿Cuál es el precio promedio de las propiedades con terraza? 

```python
sSQL = """ 

SELECT ROUND(AVG(price),2) AS avg_price_inmuebles_con_terraza
FROM inmuebles
WHERE hasterrace = 1

"""

pd.read_sql(sSQL,conn)
```

`El precio promedio de las propiedades con terraza es de 432,226.67€`


## 3. ¿Cuántas propiedades hay según el número de habitaciones? ¿Cuál es el precio medio de los apartamentos según el número de habitaciones? Ordénalo por número de habitaciones de menor a mayor

```python
sSQL = """  

SELECT roomnumber, COUNT(*) AS numero_propiedades, ROUND(AVG(price),2) AS precio_promedio
FROM inmuebles
GROUP BY roomnumber
ORDER BY roomnumber

"""

pd.read_sql(sSQL,conn)
```

## 4. ¿Cuáles son las propiedades cuyo precio es mayor que el precio promedio de todas las propiedades en el dataset? Muestra el ASSETID, PRICE y CONSTRUCTEDAREA

```python
sSQL = """  

SELECT assetid, price, constructedarea
FROM inmuebles 
WHERE price > (SELECT AVG(price) from inmuebles)

"""

pd.read_sql(sSQL,conn)
```

## 7. ¿Cuál es la distancia promedio al centro de la ciudad para las propiedades construidas después del año 2000?

Uso el año de construcción según los registros catastrales porque el año de construcción del inmueble contiene valores nulos

```python
sSQL = """ 

SELECT ROUND(AVG(distance_to_city_center),2) AS avg_distancia_centro
FROM inmuebles
WHERE cadconstructionyear > 2000

"""

pd.read_sql(sSQL,conn)
```

## 9. ¿Qué porcentaje de propiedades de segunda mano están en buen estado y tienen más de 2 baños? 

Divido las propiedades de segunda mano en buen estado con más de 2 baños, entre las propiedades de segunda mano totales(propiedades para restaurar y en buen estado).

```python
sSQL = """


SELECT
    ROUND(SUM(CASE WHEN builttypeid_3 = 1 AND bathnumber > 2 THEN 1 ELSE 0 END) * 100 / 
    COUNT(CASE WHEN builttypeid_2 = 1 OR builttypeid_3 = 1 THEN 1 END),2) AS porcentaje_inmuebles_buen_estado_con_2_baños
FROM inmuebles

"""

pd.read_sql(sSQL, conn)
```

`Un 7.61% de propiedades de segunda mano están en buen estado y tienen más de 2 baños`

## 10. ¿Hay alguna tendencia en el valor medio de las propiedades en los distintos meses del año 2018? ¿Y en el valor máximo o mínimo? 

```python
sSQL = """

SELECT period, AVG(price) AS valor_medio, MAX(price) AS valor_max, MIN(price) AS valor_min
FROM inmuebles
GROUP BY period
ORDER BY period

"""

pd.read_sql(sSQL,conn)
```

- `En cuanto al valor medio, al principio y final de año es un valor superior, a medida que se acerca el verano baja`
- `El valor máximo también tiene su máximo a final de año`
- `El valor mínimo tiene su máximo a principio de año, se reduce al entrar junio, y finalmente se recupera un poco`

## 14. ¿Cuál es el precio promedio por metro cuadrado para propiedades con orientación sur, este y oeste?

```python
sSQL = """

SELECT ROUND(AVG(unitprice),2) as avg_precio_m2
FROM inmuebles
WHERE hassouthorientation = 1 
    AND haswestorientation = 1
    AND haseastorientation = 1

"""
pd.read_sql(sSQL,conn)
```

## 15. ¿Cuál es la diferencia de los precios promedios entre propiedades que incluyen la plaza de aparcamiento y las que no? 

```python
sSQL = """
SELECT 
    AVG(CASE WHEN HASPARKINGSPACE = 1 THEN PRICE END) as parking,
    AVG(CASE WHEN HASPARKINGSPACE = 0 THEN PRICE END) as sin_parking,
    ABS(
        AVG(CASE WHEN HASPARKINGSPACE = 1 THEN PRICE END) - 
        AVG(CASE WHEN HASPARKINGSPACE = 0 THEN PRICE END)
    ) as price_difference
FROM inmuebles
"""
pd.read_sql(sSQL, conn)
```

`La diferencia en el precio de los inmuebles con parking es de 227.141,18 €`

## 1. ¿Qué propiedades gestionadas por agencias tienen piscina y plaza de aparcamiento?

## 2. ¿Cuáles son todas las propiedades gestionadas por una agencia con puntuación mayor a 4.5? Muestra el ASSETID, AGENCY_NAME, PRICE y RATING. 

```python
sSQL = """  

SELECT assetid, agency_name, price, rating
FROM inmuebles
INNER JOIN agencias
ON inmuebles.agency_id = agencias.agency_id
WHERE agencias.rating > 4.5

"""

pd.read_sql(sSQL, conn)
```

## 3. ¿Qué inmobiliarias no tienen propiedades listadas en el dataset? 

```python
sSQL = """ 

SELECT agency_name
FROM inmuebles
RIGHT JOIN agencias
ON inmuebles.agency_id = agencias.agency_id
WHERE inmuebles.agency_id IS NULL
"""

pd.read_sql(sSQL, conn)
```

`No hay ninguna inmobiliaria sin propiedades listadas`

## 4. ¿Cuál es la inmobiliaria con el precio promedio más alto en propiedades listadas? 

```python
sSQL = """ 

WITH listed_properties AS (
SELECT agency_name, AVG(price) AS avg_price
FROM inmuebles
INNER JOIN agencias
ON inmuebles.agency_id = agencias.agency_id
GROUP BY agency_name
)

SELECT agency_name
FROM listed_properties
WHERE avg_price = (SELECT MAX(avg_price) FROM listed_properties)
"""

pd.read_sql(sSQL, conn)
```

`La inmobiliaria con el precio promedio más alto es Inmo Casa Madrid`

## 5. Muestra todos los inmuebles y, si existe, el nombre de su agencia.

## 6. ¿Cuántos inmuebles gestiona cada agencia? Ordénalo de mayor a menor. 

```python
sSQL = """  

SELECT agency_name, COUNT(assetid) as viviendas_gestionadas
FROM inmuebles
INNER JOIN agencias
ON inmuebles.agency_id = agencias.agency_id
GROUP BY agency_name
ORDER BY COUNT(assetid) DESC

"""

pd.read_sql(sSQL, conn)
```
